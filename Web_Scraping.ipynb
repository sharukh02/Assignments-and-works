### Q4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
import requests
from bs4 import BeautifulSoup
page3 = requests.get("https://bookpage.com/reviews")
page3
```

```python
soup = BeautifulSoup(page3.content,"html.parser")
soup.prettify
```

```python
B_n=[]
for i in soup.find_all("div",class_="bp-well"):
    for j in i.find_all("h4",class_="italic"):
        B_n.append(j.text.replace("★","").strip())
```

```python
A_n=[]
for i in soup.find_all("div",class_="bp-well"):
    for j in i.find_all("p",class_="sans bold"):
        A_n.append(j.text.strip())
```

```python
genre=[]
for i in soup.find_all("div",class_="bp-well"):
    for j in i.find_all("p",class_="genre-links hidden-phone"):
        genre.append(j.text.replace("\n","").strip())
```

```python
url_tags = soup.find_all("h4",class_="italic")
```

```python
url = []
for i in url_tags:
    for j in i.find_all("a",href=True):
        url.append(j["href"])
url
```

```python
reviews = []
for i in url:
    book = requests.get("https://www.bookpage.com"+i)
    soup= BeautifulSoup(book.content,"html.parser")
    rev = soup.find("div",class_="article-body")
    reviews.append(rev.text)
reviews    
```

```python
import pandas as pd
df4 = pd.DataFrame({"Book Name":B_n[0:5],
                   "Author_Name":A_n[0:5],
                   "Genre":genre[0:5],
                   "Reviews":reviews[0:5]})
df4
```
