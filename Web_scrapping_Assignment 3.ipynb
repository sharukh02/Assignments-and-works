```python
while True:
    
    for i in driver.find_elements_by_xpath('//h2[@class="title title-6"]'):
        hostel.append(i.text)
    
    for i in driver.find_elements_by_xpath('//h2[@class="title title-6"]//a'):
        urls.append(i.get_attribute('href'))
    
    for i in driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]'):
        facilities.append(i.text.replace("\n",","))
    
    for i in driver.find_elements_by_xpath('//span[@class="description"]'):
        distance.append(i.text)
        
    for i in driver.find_elements_by_xpath('//div[@class="price-col"]'):
        if i.text is None:
            price.append("---")
        else:
            price.append(i.text.replace("\n"," "))
    try:
        time.sleep(2)
        driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]').click()
    
    except NoSuchElementException:
        print("No more pages to click")
        break
```

```python
description = []
overall_r=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    try:
        driver.find_element_by_xpath('//a[@class="toggle-content"]').click()# clicking on show more
    except NoSuchElementException:
        None
    
    desc = driver.find_element_by_xpath('//div[@class="collapsible"]/div[2]')
    if desc.text is None:
        description.append("---")
    else:
        description.append(desc.text)
```

```python
for i in urls:
    driver.get(i)
    try:
        rat = driver.find_element_by_xpath('//div[@class="score orange big"]')
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append("---")
        None
```

```python
for i in urls:
    driver.get(i)
    try:
        over = driver.find_element_by_xpath('//div[@class="keyword"]')
        overall_r.append(over.text)
    except NoSuchElementException:
        overall_r.append("--")
        None
```

```python
for i in urls:
    driver.get(i)    
    try:
        total = driver.find_element_by_xpath('//div[@class="reviews"]')
        if total.text is None:
            total_r.append("--")
        else:
            total_r.append(total.text)
    
    except NoSuchElementException:
        total_r.append("--")
        None    
```
